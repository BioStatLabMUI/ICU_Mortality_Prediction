## this notebook combines all the results and aggregates them 

```python
import re
from pathlib import Path

import numpy as np
import pandas as pd

# Folder containing prediction files
pred_dir = Path("./")

# Pattern:
# Fold_1_predictions_bbn_test1.csv
# Fold_1_predictions_bbn_train3.csv
pattern = re.compile(
    r"^(Fold_\d+)_predictions_(.+)_(train|test)\d+\.csv$"
)

groups = {}

# ------------------------------------------------------------------
# Group files by:
# (fold, model, split)
# ------------------------------------------------------------------
for file in pred_dir.glob("*.csv"):

    m = pattern.match(file.name)
    if not m:
        continue

    fold, model, split = m.groups()

    key = (fold, model, split)
    groups.setdefault(key, []).append(file)

# ------------------------------------------------------------------
# Average predictions within each group
# ------------------------------------------------------------------
for (fold, model, split), files in groups.items():

    print(
        f"Processing {fold} | {model} | {split} "
        f"({len(files)} imputations)"
    )

    dfs = [pd.read_csv(f) for f in sorted(files)]

    # Verify shapes match
    shapes = {df.shape for df in dfs}
    if len(shapes) != 1:
        raise ValueError(
            f"Shape mismatch in group {(fold, model, split)}: {shapes}"
        )

    # Mean prediction
    mean_df = sum(dfs) / len(dfs)

    output_name = (
        f"{fold}_predictions_{model}_{split}.csv"
    )

    mean_df.to_csv(pred_dir / output_name, index=False)

    print(f"Saved: {output_name}")
```
